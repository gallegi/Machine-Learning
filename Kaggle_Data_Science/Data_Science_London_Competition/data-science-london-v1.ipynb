```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#sklearn model
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
print(os.listdir("../input"))

# Any results you write to the current directory are saved as output.
```

```python
# read data
cols = ['col_'+str(i) for i in range(len(pd.read_csv('../input/train.csv').columns))]

train_data = pd.read_csv('../input/train.csv', names=cols)
train_target = pd.read_csv('../input/trainLabels.csv', names=['target'])

test_data = pd.read_csv('../input/test.csv', names=cols)
```

```python
train_data.head()
```

```python
train_target.head()
```

```python
# count number of null of each cols
for col in cols:
    print('Number of na of ', col, ":", train_data[col].isna().sum())
```

```python
# train logistic regression
def logistic_regression(X,y):
    model = LogisticRegression(max_iter=1000)
    model.fit(X,y)
    print(model.score(X,y))
    return model


logistic_regression(train_data, train_target)
```

```python
# train logistic regression
def svm(X,y):
    model = SVC(kernel='rbf',max_iter=1000)
    model.fit(X,y)
    print(model.score(X,y))
    return model


svm_model = svm(train_data, train_target)
```

```python
test_data.head()
```

```python
test_target = svm_model.predict(test_data)
```

```python
# submit data
submission = pd.DataFrame({'Id':np.arange(1,len(test_target)+1), 'Solution':test_target})

submission.to_csv(index=None, path_or_buf='svm_submission.csv')
```
